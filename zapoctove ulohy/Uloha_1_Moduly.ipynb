```python
from sympy import *
import scipy.integrate as integrate
import numpy as np
import time
import math

#faktoriál
#knihovna math
start = time.time()
x = 6666
result = math.factorial(x)
end = time.time()
result = str(result)
print(f"modul math:\ntime: {(end - start)}, faktorial: {result}")

#vanilla python
start = time.time()
result = 1
for i in range(1, x+1):
    result = result * i 
end = time.time()
result = str(result)
print(f"vanilla python:\ntime:{(end - start)}, faktorial: {result}\n")

#násobení matic 
#numpy
nasobek = 25
start = time.time()
matice = [[12877, 5099], [11714, 78552], [13556, 3546]]
result = np.array(matice)*nasobek
end = time.time()
print(f"numpy:\n{result}, time:{(end - start)}")

#vanilla python 
start = time.time()
matice = [[12877, 5099], [11714, 78552], [13556, 3546]]
for i in range(len(matice)):
    for j in range(len(matice[0])):
        matice[i][j] = matice[i][j]*nasobek
end = time.time()
print(f"vanilla python:\n{result}, time:{(end - start)}\n")


#skalární součin 
#numpy:
start = time.time()
a = np.array([6, 84, 75, 50])
b = np.array([12, 10, 772, 11])
result = sum(a*b)
end = time.time()
print(f"numpy: {result}, time: {(end - start)}")

#vanilla python 
start = time.time()
result = 0
a = np.array([6, 84, 75, 50])
b = np.array([12, 10, 772, 11])
for i in range(len(a)):
    result += a[i]*b[i]
end = time.time()
print(f"vanilla python: {result}, time: {(end - start)}\n")


#derivace
#sympy 
def find_deriv(f, var, hodnota):
    var = symbols("x")
    derivace = diff(f, var)
    return (derivace.subs(x, hodnota)).doit()
    
start = time.time()
x = symbols('x')
funkce = (10*x**3)*(2*x**5-5)
result = find_deriv(funkce, x, 12)
konec = time.time()
print(f"sympy derivace: {result}, time:{(konec - start)}")

#   vlastní funkce 
def f(x):
    return (10*x**3)*(2*x**5-5)

def derivace(funkce, hodnota, h=0.001):
    return (funkce(hodnota+h) - funkce(hodnota))/h
start = time.time()
result = derivace(f, 12)
end = time.time()

print(f"vanilla python: {result}, time: {(end - start)}\n")


#integrace 
#scipy
start = time.time()
result = integrate.quad(lambda x: (x**3+2*x+4)/2, 1, 4)
end = time.time()
print(f"scipy: {result[0]}, time: {(end - start)}")

#vanilla python
def f(x):
    return (x**3+2*x+4)/2

start = time.time()
result = 0
dx = 0.0001
x = 1
b = 4
while x < b:
    result += dx * (f(x) + f((x+dx)))/2
    x += dx
end = time.time()
print(f"vanilla python:{result}, time: {(end - start)}")

```
